```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
classes = ['Physics', 'Chemistry', 'Maths', 'Science', 'SST']
marks = [89, 45, 78, 23, 90]
```

```python
plt.pie(marks, labels = classes)
plt.show()
```

```python
plt.pie(marks, labels = classes,  autopct = '%0.1f%%')
plt.show()
```

```python
explode_values = [0.1, 0, 0, 0, 0]

plt.pie(marks, labels = classes, autopct = '%0.1f%%', explode=explode_values)
plt.show()
```
